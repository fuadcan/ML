# Regressions

Regression is a method to explore the relationship between variables. More precisely it measures the effect of some independent (or as it is assumed to be, exogenous) variable(s) over another independent variable. So to speak if we suspect that $X$ directly have an effect over the realizations of $Y$, we regress $X$ over $Y$  to investigate our hyptothesis.

A very basic example is the relationship between prices and the square feet measure of the houses. We know by intiution that $sqft$ is one of the main factors effecting $HP$; so in this case $Y = HP$ and $X = sqft$. Terminologically speaking we regress $X$ over $Y$; or in R notation, `Y~X`.     

## Simple Linear Regression

The first pace is simple linear regressions where there is only one independent variable, $X$, that explain the changes in $Y$. For any observation $i$:

$$Y_i = \alpha + \beta X_i + \epsilon_i$$

where $\alpha$ and $\beta$ are constants. Here $\alpha$ and $\beta$ are called _intercept_ and _slope_ of the equation and $\epsilon_i$ is _residual_. Which translates into $Y_i$ is equal to sum of some constant, a magnitude of $X_i$ and a random number $\epsilon_i$.

For observed X and Y series, of course, we don't know the $\alpha$ and $\beta$ yet. To find the best parameters, or in other words, the best line passes through the observations, we need to _estimate_ from the observations. 

### House Prices - Square Feet Example

To illustrate our problem, we will follow a very similar example used in "Coursera Machine Learning Specialization: Regressions" with some differences. We start with plotting a sample data:

```r
library("ggplot2")
options(repr.plot.width=6, repr.plot.height=4.5)
hpdata <- get(load("../data/housesalessubset.rda"))
hpdata$logprice <- log(hpdata$price)

pl <- ggplot(hpdata,aes(sqft_living,logprice,color=logprice))
pl <- pl + geom_point() + ggtitle("logHP vs sqft")
pl
```

Indeed there _seems_ a (positive) relation between two variables. While sqft_living increases price increases as well. We will now start to search for the best line that passes through the observations. (Note that the relation may not be linear but quadratic or even have a higher order. We will point this issue later.)

We will start with asking if $\bar Y$ could be the best line:

```r
pred <- rep(mean(hpdata$logprice),nrow(hpdata))

pl <- ggplot(hpdata,aes(sqft_living,logprice,color=logprice))
pl <- pl + geom_point() + ggtitle("logHP vs sqft")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

Obviously this is not the best guess that present the relationship but we usually cannot give a precise answer, _why not?_. Let's attempt another guess:

```r
pred <- predict(lm(logprice~sqft_living,hpdata)) + 1

pl <- ggplot(hpdata,aes(sqft_living,logprice,color=logprice))
pl <- pl + geom_point() + ggtitle("logHP vs sqft")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

This definitely is a better model since it captivates the behaviour of the data: as it is for the observations, for our model line, logprice increases with an increase in sqft_living increase.

Now we can talk about what is a better model, but there is still a problem here. Our model almost always give higher prices for any value of sqft_living. Now let's try another:

```r
temp <- lm(logprice~sqft_living,hpdata)

pl <- ggplot(hpdata,aes(sqft_living,logprice,color=logprice))
pl <- pl + geom_point() + ggtitle("logHP vs sqft")
pl <- pl + geom_abline(slope = coef(temp)[2] * .5,intercept = coef(temp)[1] - .7)
pl
```

```r

pl <- ggplot(hpdata,aes(sqft_living,logprice,color=logprice))
pl <- pl + geom_point() + ggtitle("logHP vs sqft")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
### Fitting the Best Line
```

Different than statistics that is a discipline aiming to theorize true relationship between variables, ML focuses on pragmatically working models. Hence it bypasses many theory of statistics and starts with regressions instead.

Regressions for ML purpose, of course, satisfy certain conditions but they are not as strict as statistics since ML leans over bigger data. In particular large sample theory implies when the number of observation inceases over a level, the obstacle against some assumptions of regressions ceases.
