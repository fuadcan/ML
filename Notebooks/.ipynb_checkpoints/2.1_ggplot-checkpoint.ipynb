__ggplot__ is a plotting library that gives a fairy touch to R. It is maybe the second best thing happened to R after the release of RStudio. See this [link](https://www.google.com.tr/search?q=before+and+after&client=ubuntu&hs=Ncf&channel=fs&dcr=0&source=lnms&tbm=isch&sa=X&ved=0ahUKEwjEx8SO6InZAhUH5qQKHbtQDGIQ_AUICigB&biw=1301&bih=668) for further details.

ggplot follows a similar logic with plot, you first plot, then add titles, lines and any other format; but here we use '+' to add layers instead of just passing to a new line. To begin with we will again import ISLR that contains Auto data.frame:

(Don't get confused with the options parameter. It just reduces the size of the plot for presenting purposes)

```r
library("ISLR")
library("ggplot2")

options(repr.plot.width=4, repr.plot.height=3) 

ggplot(Auto,aes(weight,mpg)) + geom_point() + ggtitle("MPG and Weight")
```

Here notice that "ggplot" command does not plot anything, just forms a ground for other elements. The first argument is by default data.frame to be plotted and second argument "aes" is the aesthetics which is dimensions in context of "ggplot" command.

This way of plotting gives chance to think analytically what we are plotting as every elements are added distinctly. To see it closely:

```r
pl <- ggplot(Auto,aes(weight,mpg))
pl
```

```r
pl <- pl  + geom_point()
pl
```

```r
pl <- pl  + ggtitle("MPG vs Weight")
pl
```

### Adding lines and basic formatting
As the data.frame contains all the information needed, formatting the plot is _easy-going_:

```r
## 2 ##
options(repr.plot.width=NULL, repr.plot.height=4)

pred <- predict(lm(mpg~weight+I(weight^2),data=Auto))

pl <- ggplot(Auto,aes(weight,mpg,color=horsepower))
pl <- pl + geom_point() + ggtitle("MPG ve Weight")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

As data.frame object is easy to be managed, it is just enough to extract information that has the same length with your data to add your data.frame:

```r
brand <- Auto$name
brand <- strsplit(as.character(brand)," ")
brand <- sapply(brand, function(elem) elem[[1]])
Auto$brand <- brand

## 3 ##
pred <- predict(lm(mpg~horsepower+I(horsepower^2),data=Auto))
pl <- ggplot(Auto,aes(horsepower,mpg,color=brand))
pl <- pl + geom_point() + ggtitle("MPG ve BG")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

```r
## 4 ##
options(repr.plot.width=NULL, repr.plot.height=NULL)
pl <- ggplot(Auto,aes(horsepower,mpg,color=brand))
pl <- pl + geom_text(aes(label=brand),size=2,color='black')
pl <- pl + geom_point() + ggtitle("MPG ve BG")
pl
```

```r
Auto$mpg01 <- Auto$mpg > median(Auto$mpg) # The example used in Introduction to Statistical Learning
pl <- ggplot(Auto,aes(horsepower,mpg,color=mpg01))
pl <- pl + geom_point() + ggtitle("MPG ve BG")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

```r
## 6 ##
pl <- ggplot(Auto,aes(horsepower,mpg))
pl <- pl + geom_point() + ggtitle("MPG ve BG")
pl <- pl + geom_smooth()
pl
```

```r
## 7 ##
pl <- ggplot(Auto,aes(horsepower,mpg,color=mpg01))
pl <- pl + geom_point() + ggtitle("MPG ve BG")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl <- pl + geom_smooth()
pl
```
