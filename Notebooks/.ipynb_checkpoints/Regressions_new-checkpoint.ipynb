# Regressions

Regression is a method to explore the relationship between variables. More precisely it measures the effect of some independent (or as it is assumed to be, exogenous) variable(s) over another independent variable. So to speak if we suspect that $X$ directly have an effect over the realizations of $Y$, we regress $X$ over $Y$  to investigate our hyptothesis.

A very basic example is the relationship between prices and the square feet measure of the houses. We know by intiution that $sqft$ is one of the main factors effecting $HP$; so in this case $Y = HP$ and $X = sqft$. Terminologically speaking we regress $X$ over $Y$; or in R notation, `Y~X`.     

## Simple Linear Regression

The first pace is simple linear regressions where there is only one independent variable, $X$, that explain the changes in $Y$. For any observation $i$:

$$Y_i = \alpha + \beta X_i + \epsilon_i$$

where $\alpha$ and $\beta$ are constants. Here $\alpha$ and $\beta$ are called _intercept_ and _slope_ of the equation and $\epsilon_i$ is _residual_. Which translates into $Y_i$ is equal to sum of some constant, a magnitude of $X_i$ and a random number $\epsilon_i$.

For observed X and Y series, of course, we don't know the $\alpha$ and $\beta$ yet. To find the best parameters, or in other words, the best line passes through the observations, we need to _estimate_ from the observations. 

### House Prices - Square Feet Example

To illustrate our problem, we will follow a very similar example used in "Coursera Machine Learning Specialization: Regressions" with some differences. We start with plotting a sample data:

```r
library("ggplot2")
options(repr.plot.width=6, repr.plot.height=4.5)

# Sample Data
set.seed(122)
sqft  <- seq(0,1000,50)
HP    <- 100000 + 10*sqft + 1000*rnorm(length(sqft))
HPdat <- data.frame(HP,sqft)

pl <- ggplot(HPdat,aes(sqft,HP,color=HP))
pl <- pl + geom_point() + ggtitle("HP vs sqft")
pl
```

Indeed there _seems_ a (positive) relation between two variables. While sqft_living increases price increases as well. We will now start to search for the best line that passes through the observations. (Note that the relation may not be linear but quadratic or even have a higher order. We will point this issue later.)

## An Intuitive Approach for Identifying Best Line

We will start with asking if average of $Y_i$, $\bar Y$ could be the best line:

```r
pred <- rep(mean(HPdat$HP),nrow(HPdat))

pl <- ggplot(HPdat,aes(sqft,HP,color=HP))
pl <- pl + geom_point() + ggtitle("(1) HP vs sqft")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

Obviously this is not the best guess that present the relationship but we usually cannot give a precise answer to _why not?_. Let's give another shot:

```r
temp <- lm(HP~sqft,HPdat)

pl <- ggplot(HPdat,aes(sqft,HP,color=HP))
pl <- pl + geom_point() + ggtitle("(2) HP vs sqft")
pl <- pl + geom_abline(slope = coef(temp)[2] * .5,intercept = coef(temp)[1] + 1000)
pl
```

While the new line responds to data's behavior, it still is not well enough. Now let's examine the following: 

```r
pred <- predict(lm(HP~sqft,HPdat)) + 1000

pl <- ggplot(HPdat,aes(sqft,HP,color=HP))
pl <- pl + geom_point() + ggtitle("(3) HP vs sqft")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

This definitely is a better model since it captivates the behaviour of the data by following the general pattern of the observations; as sqft_living increases, we our line as with observations, an move upward.

Now we can talk about what is a better model, but there is still a problem here. Our model almost always estimate _higher prices_ for any value of sqft_living.

__Comparison of the three models above__

(1) is not responding to any change with respect to the changes in _sqft living_. So for sqft_living < 500, the model estimates higher prices whereas for sqft_living $\geq 500$ it estimates lower prices; and this is due to the flat estimation where it does not recognizes data at all.

(2) is responding to the changes in sqft_living but it does not follow the behaviour of data.

(3) is responding to the changes in sqft_living and it is satisfactorily traces data but it has an error inside; it almost always give higher price estimations.


Now let's try another:

```r
pred <- predict(lm(HP~sqft,HPdat))

pl <- ggplot(HPdat,aes(sqft,HP,color=HP))
pl <- pl + geom_point() + ggtitle("HP vs sqft")
pl <- pl + geom_line(aes(y=pred),color='black') 
pl
```

Now this estimation looks much better; while it captures the state of the points, the estimations are almost _unbiased_.
