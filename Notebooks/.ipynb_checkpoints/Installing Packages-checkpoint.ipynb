R is in the purgatory between being a statistical software and a programming language. It was initially published as a statistical programming language however as with the recent developments in Data Science and Machine Learning, it expanded more into programming language space. Now its boundaries are unlimited; you can manage your databases, scrape the web, parse Natural Languages or even send automatic tweets. 

Relative to other languages such as Java, C and C++, its language is easy to be adapted, even for people who are agnostic to programming culture. Thus it became widespread with release of RStudio which made it basicly _aesthetic_. It now invites people from any background to design their own programmes.

However R is not that much a programming language, when you attempt to scale up your code to execute higher computations R falls short (see this [link](https://julialang.org/benchmarks/) for comparison of languages). But its community is so wide in both quantity and quality and can overcome the deficiencies by developing libraries that carry out computationsin C, C++ etc. but yield results into R. Moreover whenever a new method is developed, it does not take much to be implemented in R for free. It is the character of open-source culture where the contributers are motivated by their reputation.

## Installing Packages 

While you conduct your analysis, it is very likely for you to use external libraries, the set of programmes that does your intended work already. For example if you attempt to analyze a social network, your code written in R would probably be time inefficient since R language is not fast enough for big networks. In such a case you may choose to install and use an external library, e.g. _igraph_.

To install a library all you need is to call `install.packages()`:

```r
install.packages("igraph")
```

This will connect to CRAN to download and set the igraph package.

## Installing from GitHub 

There are times that your intended set of codes is written but is still in development in GitHub. If in the GitHub documentation it is written that you can install it via devtools library, but you need to install it first:

```r
install.packages("devtools")

devtools::install_github("hadley/reshape")
```

## Installing via RStudio

There are times when default `install.packages` doesn't work properly. You may try, then, to use RStudio's package installation that is the __install__ button on the right bottom window, under __packages__ tab:

```r
library("IRdisplay")
display_png(file="RStudio_package_installation_from_CRAN.png") 
```

This window also allows you to install from a tar.gz file. For it you first need to download the package via your favourite browser (e.g. [tar.gz file for FactoMineR package](https://cran.r-project.org/web/packages/FactoMineR/index.html), then install via RStudio:

```r
display_png(file="RStudio_package_installation_from_CRAN.png")
```
